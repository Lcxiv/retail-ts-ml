# Notebook 05: Baseline Modeling

Establishes the minimum performance benchmark using naive and statistical baselines.

```python
import pandas as pd
import numpy as np
from src.config.config import DATA_SYNTHETIC_DIR, PRIMARY_TARGET
from src.models.baseline_models import run_all_baselines
from src.evaluation.metrics import all_metrics, evaluate_by_segment
```

```python
df = pd.read_parquet(DATA_SYNTHETIC_DIR / 'synthetic_retail_sales.parquet')
cutoff = df['date'].quantile(0.8)
train_df = df[df['date'] <= cutoff]
test_df = df[df['date'] > cutoff]
```

```python
# Run baselines on total aggregated series
train_series = train_df.groupby('date')[PRIMARY_TARGET].sum()
horizon = len(test_df['date'].unique())
results = run_all_baselines(train_series, horizon)
for name, preds in results.items():
    print(f'{name}: first 5 preds =', preds[:5].round(1))
```
